```python
customers_order_df = pd.merge(
    left=order_df,
    right=customers_df,
    how='outer',
    left_on="customer_id",
    right_on="customer_id"
)
customers_order_df.head()
```

```python
payment_df = payment_df.drop(columns = ["payment_sequential","payment_installments"])
order_all_df = pd.merge(
    left=customers_order_df,
    right=payment_df,
    how='outer',
    left_on="order_id",
    right_on="order_id"
)
order_all_df.head()
```

```python
product_df = product_df.drop(columns = ["product_name_lenght","product_description_lenght", "product_photos_qty","product_weight_g","product_length_cm","product_height_cm", "product_width_cm"])
product_all_df = pd.merge(
    left=product_df,
    right=translate_product_df,
    how='outer',
    left_on="product_category_name",
    right_on="product_category_name"
)
product_all_df.head()
```

```python
orders_item_df = pd.merge(
    left=sell_df,
    right=product_all_df,
    how='left',
    left_on="product_id",
    right_on="product_id"
)
orders_item_df.head()
```

```python
orders_df=pd.merge(
    left=orders_item_df,
    right=seller_df,
    how='outer',
    left_on="seller_id",
    right_on="seller_id"
)
orders_df.head()
```

```python
review_df= review_df.drop(columns=["review_id","review_comment_title","review_comment_message","review_creation_date","review_answer_timestamp"])
orders_item_all_df = pd.merge (
    left=orders_df,
    right=review_df,
    how='left',
    left_on="order_id",
    right_on="order_id"
)
orders_item_all_df.head()
```

Visualization & Explanatory Analysis
        
       Pertanyaan 1 Bagaimana performa penjualan Perusahaan dari 2017-2018?

```python
monthly_orders_df["order_purchase_timestamp"] = pd.to_datetime(monthly_orders_df["order_purchase_timestamp"])
monthly_orders_df["bulan"] = monthly_orders_df["order_purchase_timestamp"].dt.strftime("%B-%Y")
monthly_orders_2017_2018 = monthly_orders_df[
    (monthly_orders_df["order_purchase_timestamp"].dt.year >= 2017) &
    (monthly_orders_df["order_purchase_timestamp"].dt.year <= 2018)
]

plt.figure(figsize=(10, 5))
plt.plot(monthly_orders_2017_2018["bulan"], monthly_orders_2017_2018["order_count"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Penjualan Perbulan (2017-2018)", loc="center", fontsize=20)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("Bulan", fontsize=12)
plt.ylabel("Jumlah Pejualan", fontsize=12)
plt.show()

```

           Pertanyaan 2 Persebaran Seller Dari Seller State

```python
pesebaran_seller = orders_item_all_df.groupby(by="seller_state").seller_id.nunique().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(
    x="seller_state",
    y="seller_id",
    data=pesebaran_seller,
    palette="Blues_d" 
)
plt.title("Persebaran Seller Dari Seller State", loc="center", fontsize=15)
plt.xlabel("Seller State", fontsize=12)
plt.ylabel("Number of Sellers", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

           Pertanyaan 3 Produk apa yang paling banyak dan paling sedikit terjual?

```python
sum_order_product = orders_item_all_df.groupby(by="product_category_name_english").order_id.nunique().sort_values(ascending=False).reset_index()

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))
 
colors = ["#102cd4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(x="order_id", y="product_category_name_english", data=sum_order_product.sort_values(by="order_id", ascending=False).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Kategori Produk Terlaris", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)
sns.barplot(x="order_id", y="product_category_name_english", data=sum_order_product.sort_values(by="order_id", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Kategori Dengan Sedikit Peminat", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)
 
plt.suptitle("Kategori Produk Terlaris dan Sedikit Peminat berdasarkan Total Pembelian", fontsize=20)
plt.show()
```

           Pertanyaan 4 Produk Apa saja yang mendapatkan review bagus?

```python
rating=orders_item_all_df.groupby(by=["product_category_name_english"]).agg({"review_score": "mean"}).sort_values(by="review_score")

plt.figure(figsize=(20, 5))
sns.barplot(
    x="product_category_name_english",
    y="review_score",
    data=rating,
    palette="Blues_d"
)
plt.title("Produk Dengan Rata-Rata Review Tertinggi", loc="center", fontsize=15)
plt.xlabel("Nama Produk", fontsize=12)
plt.ylabel("Rating(Mean)", fontsize=12)
plt.xticks(rotation=45, fontsize=4)
plt.yticks(fontsize=10)
plt.show()
```

           Pertanyaan 5 Metode Pembayaran Mana Yang paling Sering Dipakai?

```python
sum_payment_customers = order_all_df.groupby(by="payment_type")["payment_value"].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10, 5))

sns.barplot(
    y="payment_value", 
    x="payment_type",
    data=sum_payment_customers.sort_values(by="payment_value", ascending=False),
    hue=["#102cd4", "#D3D3D3", "#D3D3D4", "#D3D3D6", "#D5D3D3"] 
)
plt.title("Payment Value Terhadap Tiap-Tiap Tipe Transaksi", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
yticks = [1000, 2000, 4000, 8000]
ax.set_yticks(yticks)
plt.show()
```
